```python
import turtle

turtle.Screen().bgcolor("aqua")

#pensize
turtle.pensize(5)

#pencolor
turtle.pencolor("Black")

#For the sun
sunsize = 50
turtle.fillcolor("yellow")
turtle.penup()
turtle.goto(200, 150)
turtle.pendown()
turtle.begin_fill()
turtle.circle(sunsize)
turtle.end_fill()

#For the House
#Tường bên phải
turtle.fillcolor("yellow")
turtle.penup()
turtle.home()
turtle.pendown()
turtle.begin_fill()
turtle.right(90)
turtle.forward(200)
turtle.right(60)
turtle.forward(100)
turtle.right(120)
turtle.forward(200)
turtle.right(60)
turtle.forward(100)
turtle.end_fill()

#mái bên phải
turtle.fillcolor("orange")
turtle.begin_fill()
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.left(90)
turtle.forward(100)
turtle.end_fill()

#mặt trước
turtle.fillcolor("blue")
turtle.begin_fill()
turtle.right(30)
turtle.forward(200)
turtle.right(90)
turtle.forward(200)
turtle.right(90)
turtle.forward(200)
turtle.right(90)
turtle.forward(200)
turtle.end_fill()

#mái trước
turtle.fillcolor("purple")
turtle.begin_fill()
turtle.left(120)
turtle.forward(100)
turtle.left(90)
turtle.forward(175)
turtle.end_fill()

#di chuyển đến cạnh dưới mặt trước để vẽ cửa nhà
turtle.fillcolor("green")
turtle.penup()
turtle.left(60)
turtle.forward(200)
turtle.left(90)
turtle.forward(60)
turtle.pendown()

#bắt đầu vẽ cửa
turtle.begin_fill()
turtle.left(90)
turtle.forward(90)
turtle.right(90)
turtle.forward(80)
turtle.right(90)
turtle.forward(90)
turtle.end_fill()

#For the Tree
#di chuyển
turtle.penup()
turtle.left(90)
turtle.forward(300)
turtle.pendown()

#thân cây
turtle.left(90)
turtle.forward(80)

#tán cây trái
turtle.fillcolor("green")
turtle.begin_fill()
turtle.left(90)
turtle.forward(60)
turtle.right(135)
turtle.forward(120)
turtle.left(135)
turtle.forward(50)
turtle.right(135)
turtle.forward(80)
turtle.left(135)
turtle.forward(40)
turtle.right(135)
turtle.forward(60)

#tán cây phải
turtle.right(90)
turtle.forward(60)
turtle.right(135)
turtle.forward(40)
turtle.left(135)
turtle.forward(80)
turtle.right(135)
turtle.forward(50)
turtle.left(135)
turtle.forward(120)
turtle.right(135)
turtle.forward(130)
turtle.end_fill()

#hoàn thiện thân cây
turtle.fillcolor("brown")
turtle.begin_fill()
turtle.left(90)
turtle.forward(80)
turtle.left(90)
turtle.forward(70)
turtle.left(90)
turtle.forward(80)
turtle.end_fill()
```
